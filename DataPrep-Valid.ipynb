```python
# Creating new df object
data = df.copy()
```

```python
data.head()
```

```python
"""
Indicators
"""

# MA50
ma = 50
data['MA50'] = data['close'].rolling(ma).mean()

# Bollinger Bands
maB = 20
data['BB_20 ma'] = data['close'].rolling(maB).mean()
data['BB_20 sd'] = data['close'].rolling(maB).std()
data['BB_Upper Band'] = data['BB_20 ma'] + (data['BB_20 sd']*2)
data['BB_Lower Band'] = data['BB_20 ma'] - (data['BB_20 sd']*2)
data.drop(['BB_20 sd'], axis=1, inplace=True)
```

```python
print(data['MA50'].nunique())
print(data['BB_20 ma'].nunique())
print(data['BB_Upper Band'].nunique())
print(data['BB_Lower Band'].nunique())
```

```python
data.head()
```

```python
# Remove the first records which don't have MA calculation
data.drop(data.index[:ma], inplace=True)
data.head()
```

```python
from ipywidgets import FloatProgress
from IPython.display import display

# Initialize the size of the ViewPort
size_ViewPort = 60
start_ViewPort = 0
end_ViewPort = size_ViewPort

# Progress bar
f = FloatProgress(min=0, max=(len(data.index)/end_ViewPort))
f.description = 'Progress'
display(f)
# go through the DataFrame and create ViewPorts of 30 records
pic_counter = 0
while end_ViewPort <= len(data.index):
    if (round(data.iloc[end_ViewPort]['close'], 4) - round(data.iloc[end_ViewPort+1]['close'], 4)) > 0:
        data.iloc[start_ViewPort:end_ViewPort].plot(
            kind='line', grid=False, legend=False, figsize=(3, 3)).axis('off')
        PATH = f'{path}/data/images/valid/buy'
        os.chdir(PATH)
        plt.savefig('buy.' + str(pic_counter) + '.png', dpi=100)
        plt.close()
        pic_counter += 1
        start_ViewPort = end_ViewPort
        end_ViewPort += size_ViewPort
    elif (round(data.iloc[end_ViewPort]['close'], 4) - round(data.iloc[end_ViewPort+1]['close'], 4)) == 0:
        data.iloc[start_ViewPort:end_ViewPort].plot(
            kind='line', grid=False, legend=False, figsize=(3, 3)).axis('off')
        PATH = f'{path}/data/images/valid/hold'
        os.chdir(PATH)
        plt.savefig('hold.' + str(pic_counter) + '.png', dpi=100)
        plt.close()
        pic_counter += 1
        start_ViewPort = end_ViewPort
        end_ViewPort += size_ViewPort
    else:
        data.iloc[start_ViewPort:end_ViewPort].plot(
            kind='line', grid=False, legend=False, figsize=(3, 3)).axis('off')
        PATH = f'{path}/data/images/valid/sell'
        os.chdir(PATH)
        plt.savefig('sell.' + str(pic_counter) + '.png', dpi=100)
        plt.close()
        pic_counter += 1
        start_ViewPort = end_ViewPort
        end_ViewPort += size_ViewPort
    f.value += 1
f.bar_style = 'success'
```
